# Mengubah start_time and end_time to starttime and endtime

```python
folder_path = "tostarttimeandendtime"

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        df = pd.read_csv(file_path)
        diubah = False

        if "start_time" in df.columns:
            df.rename(columns={"start_time": "starttime"}, inplace=True)
            diubah = True
        if "end_time" in df.columns:
            df.rename(columns={"end_time": "endtime"}, inplace=True)
            diubah = True
            
        if diubah:
            df.to_csv(file_path, index=False)
            print(f"Berhasil mengubah kolom di file: {filename}")
```

```python
data_folder = "trip/2013-to-2020/"
print(data_folder)
df = []

for file in os.listdir(data_folder):
    if file.endswith('.csv'): 
        print('loading file {0}...'.format(file))
        current_df = pd.read_csv(os.path.join(data_folder, file))
        
        current_df.rename(columns={'endtime': 'stoptime'}, inplace=True)
        
        if 'birthday' in current_df.columns:
            current_df.drop(columns=['birthday'], inplace=True)
        df.append(current_df)

df_master = pd.concat(df, ignore_index=True)
df_master.to_csv("2013-2020.csv", index=False)
```
